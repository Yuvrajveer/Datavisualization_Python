###**Dropping Column**
* Dropping unnecessary column which is not required for analysis.




```python
df.drop(['Sales Channel'],axis=1,inplace=True)
```

```python
df.head(5)
```

###**EDA**

```python
df['Order Priority'].value_counts()
sns.countplot(x='Order Priority',data=df)
plt.rcParams['figure.dpi']=100
plt.rcParams['figure.figsize']=(5,5)
```

```python
df['Total Profit'].describe()
```

###**Advanced Insights**

###**1. How many units sold region wise**



```python
Sold_units=df.groupby('Region')['Units Sold'].sum()
Sold_units.plot(kind='bar',color='g')
plt.rcParams['figure.figsize']=(5,5)
plt.title('Units Sold Region Wise')
plt.xlabel('Region')
plt.ylabel('Units Sold')
```

###**Total Profit Region Wise**

```python
profit=df.groupby('Region')['Total Profit'].sum()
profit.plot(kind='bar',color='r')
plt.rcParams['figure.figsize']=(5,5)
plt.title('Total Profit Region Wise')
plt.xlabel('Region')
```

###**Units Sold Item Wise**

```python
Items=df.groupby('Item Type')['Units Sold'].sum()
Items.plot(kind='pie',autopct='%1.1f%%')
plt.rcParams['figure.figsize']=(5,5)
plt.title('Units Sold Item Wise')
```
